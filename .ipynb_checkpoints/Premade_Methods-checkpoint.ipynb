```python
import numpy as np
import matplotlib.pyplot as plt
import os
import sys
import cv2
```

```python
# Import Classifier for Face and Eye Detection
face = cv2.CascadeClassifier('frontface.xml')
def face_detector (image):
    # Convert Image to Grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    faces = face.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))
    
    print("Found {0} faces!".format(len(faces)))

    # Draw a rectangle around the faces
    for (x, y, w, h) in faces:
        cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 2)
    return image
```

```python
folder = './data/'
pics = os.listdir(folder)

idx = 368

plt.imshow(cv2.cvtColor(face_detector(cv2.imread(folder + pics[idx])), cv2.COLOR_BGR2RGB))
```

```python
from scipy import ndimage as ndi

from skimage import feature

im = cv2.cvtColor(cv2.imread(folder + pics[idx]), cv2.COLOR_BGR2GRAY)

# Compute the Canny filter for two values of sigma
edges1 = feature.canny(im)
edges2 = feature.canny(im, sigma=3)

# display results
fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3, figsize=(20, 20),
                                    sharex=True, sharey=True)
ax1.imshow(im, cmap=plt.cm.gray)
ax1.axis('off')
ax1.set_title('noisy image', fontsize=20)

ax2.imshow(edges1, cmap=plt.cm.gray)
ax2.axis('off')
ax2.set_title('Canny filter, $\sigma=1$', fontsize=20)

ax3.imshow(edges2, cmap=plt.cm.gray)
ax3.axis('off')
ax3.set_title('Canny filter, $\sigma=3$', fontsize=20)

fig.tight_layout()

plt.show()
```
