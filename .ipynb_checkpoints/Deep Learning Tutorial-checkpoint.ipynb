```python
import keras
from keras.models import Model
from keras.layers import Input, Conv2D, Dense, Flatten, Reshape, MaxPooling2D, Flatten, Reshape
from keras.layers import BatchNormalization, GaussianNoise, Dropout
from keras.optimizers import Adam, RMSprop
from keras import losses, metrics
from keras.callbacks import LambdaCallback, EarlyStopping, ModelCheckpoint
import keras.backend as K
import sys
import os
import numpy as np
```

```python
valid_split = 0.1
from keras.datasets import mnist
from keras.utils import to_categorical

(X, Y), (X_test, Y_test) = mnist.load_data()

# When using a NN, use 28*28
# When using a CNN, you need a 3D model so you use 28,28,1.
# If you're using RGB, use 28x28x3
X = X.reshape(60000,28*28,)
X_test = X_test.reshape(10000,28*28)
Y = to_categorical(Y)
Y_test = to_categorical(Y_test)

print(X.shape, Y.shape)
print(X_test.shape, Y_test.shape)

idx = np.random.shuffle(np.array(range(X.shape[0])))
valid_amount = int(X.shape[0] * valid_split)

X_train = X[valid_amount:]
Y_train = Y[valid_amount:]
X_valid = X[:valid_amount]
Y_valid = Y[:valid_amount]

print("=================================")
print(X_train.shape, Y_train.shape)
print(X_valid.shape, Y_valid.shape)
print(X_test.shape, Y_test.shape)
```

```python
# Model Initialization
K.clear_session()  #clear any previous model parts

def architecture(input):
    #architecture of the network
    layers = [] 
    
    # CNN example
    # layers += [Conv2D(4, 5, strides=2, padding='same', activation='relu')]
    # layers += [Conv2D(8, 5, strides=2, padding='same', activation='relu')]
    # layers += [Flatten()]
    # layers += [Dense(64, activation='sigmoid')]
    # layers += [Dense(10, activation='softmax')]
    
    # NN example
    layers += [Dense(256, activation='relu')]
    layers += [Dense(32, activation='sigmoid')]
    layers += [Dense(10, activation='softmax')]

    network = input
    for layer in layers:
        network = layer(network)

    network = Model(input, network, name='Neural Network') # model instantiation
    return network

inputs = Input(shape=(28*28,), name='Input')
model = architecture(inputs)
model.compile(loss='binary_crossentropy', optimizer='RMSprop')
model.summary()
```

```python
earlystop = EarlyStopping(monitor='val_loss', min_delta=0, patience=5, verbose=0, mode='auto')
checkpoint = ModelCheckpoint('test.h5', monitor='val_loss', verbose=0, save_best_only=True)

try:
    model.fit(x=X_train, y=Y_train, validation_data=(X_valid, Y_valid), epochs=1000, batch_size=32, verbose=1, callbacks=[earlystop, checkpoint])
except KeyboardInterrupt:
    pass
```

```python
import matplotlib.pyplot as plt

Y_pred = model.predict(X_test)

idx = 10
print(np.argmax(Y_test[idx]))
print((Y_pred[idx] * 100).astype(int))

plt.imshow(X_test[idx].reshape(28,28), cmap='gray')
```

```python
n = 100
ncol = 10
nrow = int((n + ncol - 1) / ncol)

fig, plots = plt.subplots(nrow, ncol, figsize=(2*ncol,2*nrow))
plots = plots.flatten()

for i in range(n):
    p = X_test[i].reshape(28,28)
    pred = np.argmax(Y_pred[i])
    true = np.argmax(Y_test[i])
    plots[i].set_xticks([])
    plots[i].set_yticks([])
    
    if pred == true:
        plots[i].set_title(pred, color='black', fontsize=20, fontweight='bold')
    else:
        plots[i].set_title(pred, color='red', fontsize=20, fontweight='bold')
    plots[i].imshow(p, cmap="gray")
```
