```python
import os
import numpy as np
import pandas as pd
import sys
import urllib.request
```

```python
df = pd.read_excel('20190111_minted_2012_holiday_challenge_design_submission_photo_attributes.xlsx')
df.columns = ['id', 'url', 'pic_amount']
df['pic_amount'] = df['pic_amount'].replace('6+', 6)
df
```

```python
folder = './data/'
def download(x):
    index, url, num = x
    urllib.request.urlretrieve(url, folder + str(index) + '.' + str(num) + ".png")
    
df.apply(download, axis=1)
```
